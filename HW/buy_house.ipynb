```python
import pandas as pd

# Предположим, что df - это ваш DataFrame
# Преобразование столбца Date в формат datetime
df['Date'] = pd.to_datetime(df['Date'])

# Создаем столбец Quarter с использованием strftime для уникального представления квартала
df['Quarter'] = df['Date'].dt.strftime('%YQ%q')

# Подсчет количества проданных объектов в каждом квартале
quarterly_sales = df.groupby('Quarter').size()

# Находим второй по популярности квартал
second_most_popular_quarters = quarterly_sales.sort_values(ascending=False).index[1]

# Получаем количество проданных объектов во втором по популярности квартале
sales_in_second_quarter = quarterly_sales.loc[second_most_popular_quarters]

# Выводим результат
print(f"Второй по популярности квартал: {second_most_popular_quarters}, Продажи: {sales_in_second_quarter}")
```
