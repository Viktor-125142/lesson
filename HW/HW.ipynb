```python
import time
import pandas as pd
import requests
from bs4 import BeautifulSoup

URL = "https://www.banki.ru/banks/ratings/?PAGEN_1="

list_of_dfs = []

# Цикл по страницам
for i in range(1, 9):
    # Отправка GET-запроса
    response = requests.get(URL + str(i))

    # Проверка успешности запроса
    if response.status_code == 200:
        # Используем BeautifulSoup для парсинга HTML
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Находим HTML-таблицу
        table = soup.find('table')

        # Чтение HTML-таблицы с использованием pd.read_html
        list_df_page = pd.read_html(str(table), encoding="cp1251")
        
        # Добавление DataFrame в список
        list_of_dfs.append(list_df_page[0])

        # Пауза для предотвращения чрезмерного обращения к серверу
        time.sleep(1)
    else:
        print(f"Error: {response.status_code}")

# Объединение всех DataFrame в один
DF = pd.concat(list_of_dfs, ignore_index=True)

# Вывод результирующего DataFrame
print(DF)
```
